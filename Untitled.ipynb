```python
import pandas as pd
import numpy as np
import warnings as w

w.filterwarnings('ignore')

def movie_recommend(movie_selected='Star Wars (1977)'):
    
    #readfiles
    cols=['user_id','movie_id','rating','tsp']
    a=pd.read_csv('u.data',sep='\t',names=cols)
    item_cols=['movie_id','title']+[str(i) for i in range (22)]
    b=pd.read_csv('u.item',sep='|',encoding='ISO-8859-1',names=item_cols)[['movie_id','title']]
    
    #merge the dataframe
    #m=merged dataset
    m=pd.merge(a,b,on='movie_id')
    
    #creating the pivot table
    #q=pivot table
    q=m.pivot_table(index='user_id',columns='title',values='rating')
    
    #correlation stuff
    corrs=q.corrwith(q[movie_selected])
    df=pd.DataFrame(corrs,columns=['correlation'])
    df['rating']=(m.groupby('title')['rating'].mean().sort_values(ascending=False))
    df['count']=m['title'].value_counts()
    
    #top3reccomedation
    t3r=list(df[df['count']>50].sort_values(by='correlation',ascending=False).head(3).index)
    if movie_selected in t3r:
        t3r.remove(movie_selected)
    
    return(t3r)
```

```python
movie_recommend('8 Seconds (1994)')
```
